```python
import pandas as pd

dataset = 'amazon_sales_report.csv'

df = pd.read_csv(dataset)

# change data format to data
df['Date'] = pd.to_datetime(df['Date'])

# Cleaning up column names
df.columns = df.columns.str.strip().str.lower().str.replace(' ','_').str.replace('-', '_').str.replace(':', '')
```

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
# # Drop useless columns
df.drop(columns=['unnamed_22', 'fulfilled_by'], inplace=True)

# Promotion flag
df['promotion_used'] = df['promotion_ids'].notnull().astype(int)

# Drop missing revenue
df = df.dropna(subset=['amount'])

# Handle currency
df.drop(columns=['currency'], inplace=True)

# Fill categorical nulls
df['courier_status'].fillna('Unknown', inplace=True)
df['ship_city'].fillna('Unknown', inplace=True)
df['ship_state'].fillna('Unknown', inplace=True)
df['ship_country'].fillna('Unknown', inplace=True)
df['ship_postal_code'].fillna('Unknown', inplace=True)
```

```python
df['order_month'] = df['date'].dt.to_period('M')
df['is_delivered'] = df['status'].apply(lambda x : 1 if "Delivered" in x else 0)
df['revenue'] = df['qty'] * df['amount']
```

```python
# Exporting to new csv 
df.to_csv("cleaned_data.csv")
```
